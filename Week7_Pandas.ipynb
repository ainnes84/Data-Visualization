## Set Index
- Column named 'State' is set as the index column.

```python
df1 = data.set_index('State')

df1
```

## Group All Rows Accordingly to State
- We can aggregate and ask for the sum total of State.  Or aggregate and ask for the mean.
- The function below aggregates all the data that corresponds to each State into one single row per State.  For example, if AZ has 256 rows out of 5,413 total rows, now it will all show as one mean of all the data into one single row.  Dataframe has 5,413 rows and we get 26 total rows after running the function below.

```python
df2 = df1.groupby('State').mean()

df2
```

## Second Way of Aggregating
- Below is the example of grouping and asking for the sum instead of the mean().

```python
df2 = df1.groupby('State').sum()

df2
```

## Select Desired Columns
- The columns 'Branch Number', 'Zipcode', 'Latitude',and  'Longitude' are not useful anymore.  All the zipcodes are sumed up and meaningless to use.  Same for the mentioned columns as well.  But we can drop and get rid of those columns and keep what we need using the below example.

```python
df3 = df2.iloc[:, 5:]

df3
```

## Note from the Above Example
- The first : inside square brackets denotes all the rows, and the second : after number 5 denotes all the columns after the fifth column.  So, instead we can try [:, 5:6] which means all the rows and only the fifth and sixth column are to be selected from the entire dataframe.
- Same goes for the rows.  So, at the first : we can place a number in front or after like this [4:, 5:] which will return all the rows after the fourth row (excluding the first four).  Or the opposite [:4, 5:] which will return only the first 4 rows.

## Method describe()

## Get Rid of Rows
- If you look at the above data, rows that correspond to 'DC', 'PA', & 'MA' have 0 values.  We can get rid of those rows and have a better data frame to use by doing the following:

```python
df4 = df3.drop(['DC', 'MA', 'PA'])
df4
```

## Rank Data (Ascending Order)
- Rank the data in ascending order.  The example below lists the smallest number first by specifying (..ascending=True).  We get the same result using (...ascending=1).

```python
ranked1 = df4.sort_values('2010 Deposits', ascending=True)
ranked1
```

## Rank Data (Descending Order)
- Rank the data in descending order.  The example below lists the biggest number first by specifying (...ascending=False) same as (...ascending=0).

```python
ranked2 = df4.sort_values('2011 Deposits', ascending=False)
ranked2
```

## Add Column
- We can add a column in the dataframe and also rank it.

```python
df4['rank_column'] = df4['2010 Deposits'].rank(ascending=False)
data
```
